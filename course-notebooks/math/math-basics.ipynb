```python
import math

input = 2
result = math.sqrt(input)
print(f"result:, {result}")

# absolute error
absolute_error = (result ** 2 - 2)
print(f"absolute_error:, {absolute_error}")


# relative error
relative_error = absolute_error / input
print(f"relative_error:, {relative_error}")
```
